**This cord determines the slope of the RP1 frequency until it reachs RP1 burst**

This cord was used for the Figure 14I in the manuscript.
Picture1.png

```python
#get local time 
x = []      # x is the cumulative time in second
y = []      # manually input y which are the values for each spike frequency: y values have to be separated by comma

length = len(y)

print("y =", y)
print("y length =", length)

for i in range(length):
    x.append(1/(y[i])) 
print("Local time x =", x)


# cumulative x
cum_x = []

for i in range(length):
  if i == 0:
    cum_x.append(x[0])
  else:
    cum_x.append(cum_x[i-1] + x[i])

 
print("cumulative time cum_x =", cum_x)


# get local slope

z = []

for i in range(length - 1):
  z.append((y[i+1]-y[i])/(x[i+1]))
print("local slope z =", z)


z1 = []
for i in range(length - 1):
  if z[i] < 0.05:                         #0.05 is the cut off used for RP1 burst
    z1.append(z[i])
  else:
    break
print("Thre. local slope z =", z1)


#get the number of spike freq after removing RP1 burst 

len_z1 = len(z1)
y1 = []

for i in range(len_z1 + 1):
  y1.append(y[i])

print("Thre y (y1)=", y1)


## get slope for that RP1 period

#get cumulative time

print("Local time x =", x)

x1 = []

length = len(y1)

print("y1 length =", length)

for i in range(length):
  if i == 0:
    x1.append(x[0])
  else:
    x1.append(x1[i-1] + x[i])

 
print("thre cumulative time x1 =", x1)




#calculating slope
import numpy as np
import matplotlib.pyplot as plt


#get graph of all scatter as a reference

cum_x = np.array(cum_x)
y = np.array(y)

#Fit line
slope, intercept = np.polyfit(cum_x, y, 1)

#Plot
plt.figure()
plt.scatter(cum_x, y) 
plt.plot(np.unique(cum_x), np.poly1d(np.polyfit(cum_x, y, 1))(np.unique(cum_x)), color = 'k')
plt.show()

print(slope)



#Make x1 and y1 usable to numpy array

x1 = np.array(x1)
y1 = np.array(y1)

#Fit line
slope, intercept = np.polyfit(x1, y1, 1)

#Plot
plt.figure()
plt.scatter(x1, y1) 
plt.plot(np.unique(x1), np.poly1d(np.polyfit(x1, y1, 1))(np.unique(x1)), color = 'k')
plt.show()

print(slope)
```
